```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def values_frequency(df, column, color="blue"):
    try:
        fig = plt.figure(figsize = (15,10))
        ax = fig.gca()
        df[column].value_counts().plot(ax=ax, kind='bar', color=color, edgecolor='black', title=f"Frequency values for column {column}")
    except Exception as ex:
        print(ex)
```

# Exploraci√≥n de datos
**Antes de ponernos a crear un modelo o nuevas variables, vamos a ver con que tipo de datos nos encontramos y ver cuan significativos pueden ser nuestros features.**

```python
train = pd.read_csv("../data/train.csv")
test = pd.read_csv("../data/test.csv")

print(train.shape)
print(test.shape)
```

```python
print(train.info())
train.head()
```

```python
print(test.info())
test.head()
```

**Podemos plotear en gr√°ficas la frecuencia de diferentes features, por ejemplo, para empezar, podemos mostrar la frecuencia en la columna de categor√≠a, l√≠quido/no_l√≠quido, frecuencia de categorias l√≠quidas/frecuencia de categor√≠as l√≠quidas...**

```python
values_frequency(train, "category")
values_frequency(test, "category")
```

**Tambi√©n podemos visualizar para otras variables....**

```python
values_frequency(train, "is_liquid", color=["lightgreen", "lightblue"])
```

**O ir m√°s all√° y filtrar por valores de la columna en si...**

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10), frameon=False)
train.loc[train["is_liquid"] == 0]['category'].value_counts().plot(ax=ax1, kind='bar', edgecolor='black', title="No liquids frequency")
train.loc[train["is_liquid"] == 1]['category'].value_counts().plot(ax=ax2, kind='bar', edgecolor='black', title="Liquids frequency")
```

**Haciendo este peque√±a exploraci√≥n nos permite analizar que tipo de datos nos llegan y nos ayudan a pensar en posibles features adicionales a partir de estos, como podr√≠a ser un feature adicional llamado is_not_liquid...**

**Por otra parte tambi√©n podemos analizar el comportamiento m√°s estad√≠stico de los datos y ver si hay outliers. Nosotros asumiremos de primeras que los valores que nos dan son correctos, ya que no sabemos los alimentos al fin y al cabo, y se pueden dar situaciones con valores nutricionales altos respecto otros.**

```python
train.select_dtypes("float64").hist(figsize=(16, 20), bins=50, xlabelsize=8, ylabelsize=8)
```

**Otro punto a tener en cuenta dentro de nuestro proceso de exploraci√≥n es ver la correlaci√≥n entre variables. Como podemos apreciar hay bastante correlaci√≥n entre valores nutricionales como era de esperar!**
